```python
# normalize date column
df1['incident_date'] = pd.to_datetime(df1['incident_date'], format = '%Y-%m-%d')

# filter out only the columns we want and drop rows with NaN values for lat and lon

filtered_df = df1.filter(items = ["incident_date",
                                   "incident_category",
                                   "incident_subcategory",
                                   "incident_description",
                                   "incident_day_of_week",
                                   "Resolution",
                                   "incident_code",
                                   "incident_id", 
                                   "incident_number", 
                                   "latitude", 
                                   "longitude"])
clean_df1 = filtered_df.dropna()

clean_df1
```

```python
# select only the dates wanted

start_date = '2019-01-1'
end_date = '2019-04-10'

mask = (clean_df1['incident_date'] > start_date) & (clean_df1['incident_date'] <= end_date)

clean_df1 = clean_df1.loc[mask]
clean_df1
```
